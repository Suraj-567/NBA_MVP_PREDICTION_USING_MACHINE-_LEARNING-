```python
import pandas as pd

# Replace 'your_file.csv' with the name of your input CSV file.
input_filename = 'player_mvp_stats.csv'

# Replace 'output_file.csv' with the name you want for the new, shorter file.
output_filename = 'player_mvp_stats.csv'

# Read the CSV into a pandas DataFrame.
# header=None tells pandas that your file doesn't have a header row.
df = pd.read_csv(input_filename, header=None)

# Keep only the first 16,248 rows (indices 0 through 16247)
df_shortened = df.iloc[:16250]

# Save the shortened DataFrame to a new CSV file.
# index=False prevents pandas from writing the row numbers as a new column.
# header=False ensures no header is written to the new file.
df_shortened.to_csv(output_filename, index=False, header=False)

print(f"Successfully created '{output_filename}' with the first {len(df_shortened)} rows.")
```

```python
import pandas as pd

df = pd.read_csv("player_mvp_stats.csv")
df.isna().any().any()

```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Awards'])

# Fill shooting % NaNs
df[['FG%', '2P%', 'eFG%']] = df[['FG%', '2P%', 'eFG%']].fillna(0)

# Fill 3P% NaNs
df['3P%'] = df['3P%'].fillna(0)

# Fill FT% NaNs using year-wise median (best option)
df['FT%'] = df.groupby('Year')['FT%'].transform(lambda x: x.fillna(x.median()))
```
