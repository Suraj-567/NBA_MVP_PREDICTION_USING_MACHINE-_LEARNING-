```python
# ===============================
# üèÄ NBA MVP Prediction: Data Cleaning & Preprocessing
# ===============================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# -------------------------------
# 1Ô∏è‚É£ Load the dataset
# -------------------------------
df = pd.read_csv("player_mvp_stats.csv")

print("Initial shape:", df.shape)
print("Columns:", df.columns.tolist()[:10], "...")  # preview first few column names


# -------------------------------
# 2Ô∏è‚É£ Drop irrelevant or leakage columns
# -------------------------------
drop_cols = ['Unnamed: 0', 'Share', 'Pts Won', 'Pts Max']
df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True, errors='ignore')

print("After dropping leakage columns:", df.shape)


# -------------------------------
# 3Ô∏è‚É£ Handle missing values
# -------------------------------
# Check how many missing values per column
print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Fill numeric missing values with column mean
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# If categorical columns exist, fill with mode (optional)
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])


# -------------------------------
# 4Ô∏è‚É£ Feature engineering
# -------------------------------
# Ensure REB column exists
if 'REB' not in df.columns:
    # Create REB from TRB (or ORB + DRB)
    if 'TRB' in df.columns:
        df['REB'] = df['TRB']
    elif 'ORB' in df.columns and 'DRB' in df.columns:
        df['REB'] = df['ORB'] + df['DRB']
    else:
        raise ValueError("No TRB or (ORB + DRB) columns found to create REB feature.")

# Sanity check
print("\nREB feature preview:")
print(df[['Player', 'Year', 'REB']].head())


# -------------------------------
# 5Ô∏è‚É£ Select core features
# -------------------------------
feature_cols = ['G', 'MP', 'FG%', '3P%', 'FT%', 'REB', 'AST', 'STL', 'BLK', 'PTS']
target_col = 'is_mvp'

# Filter only necessary columns for modeling
df_model = df[feature_cols + ['Year', target_col, 'Player']].copy()

print("\nData subset shape:", df_model.shape)


# -------------------------------
# 6Ô∏è‚É£ Split dataset by Year
# -------------------------------
train_df = df_model[df_model['Year'] <= 2015]
val_df   = df_model[(df_model['Year'] >= 2016) & (df_model['Year'] <= 2024)]
test_df  = df_model[df_model['Year'] == 2025]

print("\nTrain/Val/Test split sizes:")
print("Train:", train_df.shape, "Val:", val_df.shape, "Test:", test_df.shape)


# -------------------------------
# 7Ô∏è‚É£ Separate features (X) and labels (y)
# -------------------------------
X_train, y_train = train_df[feature_cols], train_df[target_col]
X_val, y_val     = val_df[feature_cols], val_df[target_col]
X_test, y_test   = test_df[feature_cols], test_df[target_col]  # test labels are optional (for 2025 predictions)


# -------------------------------
# 8Ô∏è‚É£ Normalize / Scale numeric features
# -------------------------------
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

print("\nFeature scaling complete!")
print("Train set mean (after scaling):", X_train_scaled.mean().round(3))
print("Train set std (after scaling):", X_train_scaled.std().round(3))
```

```python
# ===============================
# ‚öôÔ∏è Model Training Workflow Setup
# ===============================

import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression, RidgeClassifierCV
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
)
from sklearn.preprocessing import StandardScaler

# Helper function: Evaluate model
def evaluate_model(y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    auc = roc_auc_score(y_true, y_prob)
    print(f"Accuracy: {acc:.3f} | Precision: {prec:.3f} | Recall: {rec:.3f} | F1: {f1:.3f} | ROC-AUC: {auc:.3f}")
    return acc, prec, rec, f1, auc


# Helper function: Pure MVP prediction (top players overall)
def pure_mvp_predictions(val_df, y_prob, top_k=5):
    val_df_copy = val_df.copy()
    val_df_copy['mvp_prob'] = y_prob
    top_players = val_df_copy.sort_values('mvp_prob', ascending=False).head(top_k)
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top_players[['Year', 'Player', 'mvp_prob', 'is_mvp']])
    correct = top_players['is_mvp'].sum()
    print(f"‚úÖ Pure MVP Accuracy (Top-{top_k}): {correct}/{top_k} = {correct/top_k:.2f}")
    return top_players


# Helper function: Seasonal MVP prediction (per year)
def seasonal_mvp_predictions(val_df, y_prob):
    val_df_copy = val_df.copy()
    val_df_copy['mvp_prob'] = y_prob
    correct = 0
    total_years = len(val_df_copy['Year'].unique())

    print("\nüèÄ Seasonal MVP Predictions by Year:")
    results = []
    for year, group in val_df_copy.groupby('Year'):
        top_player = group.loc[group['mvp_prob'].idxmax()]
        results.append([year, top_player['Player'], top_player['mvp_prob'], int(top_player['is_mvp'])])
        if top_player['is_mvp'] == 1:
            correct += 1

    acc = correct / total_years
    results_df = pd.DataFrame(results, columns=['Year', 'Predicted MVP', 'Predicted Prob', 'Actual MVP?'])
    print(results_df)
    print(f"\n‚úÖ Seasonal MVP Accuracy: {correct}/{total_years} = {acc:.2%}")
    return results_df, acc
```

```python
# ===============================
# üîπ Logistic Regression
# ===============================
print("\n===============================")
print("üîπ Logistic Regression Model")
print("===============================")

log_model = LogisticRegression(
    solver='liblinear',
    C=1.0,
    class_weight='balanced',
    random_state=42
)

# Train
log_model.fit(X_train_scaled, y_train)

# Predict
y_val_pred = log_model.predict(X_val_scaled)
y_val_prob = log_model.predict_proba(X_val_scaled)[:, 1]

# Evaluate
evaluate_model(y_val, y_val_pred, y_val_prob)

# MVP predictions
pure_mvp_predictions(val_df, y_val_prob)
seasonal_mvp_predictions(val_df, y_val_prob)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# --- Evaluation function (same as before) ---
def evaluate_classification_model(model, X_val_scaled, y_val, df_val, model_name):
    y_pred = model.predict(X_val_scaled)
    y_proba = model.predict_proba(X_val_scaled)[:, 1]

    print(f"\nüìä Evaluation Report for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_val, y_pred):.3f} | "
          f"Precision: {precision_score(y_val, y_pred, zero_division=0):.3f} | "
          f"Recall: {recall_score(y_val, y_pred):.3f} | "
          f"F1: {f1_score(y_val, y_pred):.3f} | "
          f"ROC-AUC: {roc_auc_score(y_val, y_proba):.3f}")

    df_val = df_val.copy()
    df_val["mvp_prob"] = y_proba
    df_val["pred"] = y_pred

    # --- Pure MVP (Top 5 overall) ---
    top5 = df_val.sort_values("mvp_prob", ascending=False).head(5)
    correct_top5 = top5["is_mvp"].sum()
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top5[["Year", "Player", "mvp_prob", "is_mvp"]])
    print(f"‚úÖ Pure MVP Accuracy (Top-5): {correct_top5}/5 = {correct_top5/5:.2f}")

    # --- Seasonal MVP (per year) ---
    seasonal_preds = []
    for year, group in df_val.groupby("Year"):
        top_player = group.sort_values("mvp_prob", ascending=False).iloc[0]
        seasonal_preds.append({
            "Year": year,
            "Predicted MVP": top_player["Player"],
            "Predicted Prob": top_player["mvp_prob"],
            "Actual MVP?": top_player["is_mvp"]
        })

    seasonal_df = pd.DataFrame(seasonal_preds)
    seasonal_acc = seasonal_df["Actual MVP?"].sum() / len(seasonal_df)

    print("\nüèÄ Seasonal MVP Predictions by Year:")
    print(seasonal_df)
    print(f"\n‚úÖ Seasonal MVP Accuracy: {seasonal_acc*100:.2f}%")

    return seasonal_df, seasonal_acc


# --- Smarter Logistic Regression Tuning ---
log_reg = LogisticRegression(max_iter=5000, penalty='l2')

param_dist = {
    'C': np.logspace(-3, 2, 20),   # finer granularity
    'solver': ['liblinear', 'lbfgs'],
    'class_weight': ['balanced', None]
}

random_log = RandomizedSearchCV(
    log_reg,
    param_distributions=param_dist,
    n_iter=30,
    scoring='average_precision',  # better for rare positive class
    cv=5,
    random_state=42,
    verbose=2,
    n_jobs=-1
)

random_log.fit(X_train_scaled, y_train)
best_log = random_log.best_estimator_
print("Best Logistic Regression Params:", random_log.best_params_)

# --- Calibrate probabilities ---
calibrated_log = CalibratedClassifierCV(best_log, method='sigmoid', cv=5)
calibrated_log.fit(X_train_scaled, y_train)

# --- Evaluate ---
seasonal_df_log, seasonal_acc_log = evaluate_classification_model(
    calibrated_log, X_val_scaled, y_val, df_val, "Logistic Regression (Final Optimized)"
)
```

```python
import numpy as np
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
)

# --- Helper Function for Ridge Evaluation ---
def evaluate_ridge_model(model, X_val_scaled, y_val, df_val, model_name):
    """
    Evaluate RidgeClassifier model using decision_function scores
    (converted to 0‚Äì1 probability-like values)
    """
    y_pred = model.predict(X_val_scaled)
    
    # RidgeClassifier has no predict_proba, use decision_function
    scores = model.decision_function(X_val_scaled)
    
    # Normalize to 0‚Äì1 range for ROC-AUC & ranking
    y_proba = (scores - scores.min()) / (scores.max() - scores.min())
    
    print(f"\nüìä Evaluation Report for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_val, y_pred):.3f} | "
          f"Precision: {precision_score(y_val, y_pred, zero_division=0):.3f} | "
          f"Recall: {recall_score(y_val, y_pred):.3f} | "
          f"F1: {f1_score(y_val, y_pred):.3f} | "
          f"ROC-AUC: {roc_auc_score(y_val, y_proba):.3f}")
    
    # Attach probabilities to df for MVP ranking
    df_val = df_val.copy()
    df_val["mvp_prob"] = y_proba
    df_val["pred"] = y_pred
    
    # --- Pure MVP (Top 5 overall) ---
    top5 = df_val.sort_values("mvp_prob", ascending=False).head(5)
    correct_top5 = top5["is_mvp"].sum()
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top5[["Year", "Player", "mvp_prob", "is_mvp"]])
    print(f"‚úÖ Pure MVP Accuracy (Top-5): {correct_top5}/5 = {correct_top5/5:.2f}")
    
    # --- Seasonal MVP (per year) ---
    seasonal_preds = []
    for year, group in df_val.groupby("Year"):
        top_player = group.sort_values("mvp_prob", ascending=False).iloc[0]
        seasonal_preds.append({
            "Year": year,
            "Predicted MVP": top_player["Player"],
            "Predicted Prob": top_player["mvp_prob"],
            "Actual MVP?": top_player["is_mvp"]
        })
    
    import pandas as pd
    seasonal_df = pd.DataFrame(seasonal_preds)
    seasonal_acc = seasonal_df["Actual MVP?"].sum() / len(seasonal_df)
    
    print("\nüèÄ Seasonal MVP Predictions by Year:")
    print(seasonal_df)
    print(f"\n‚úÖ Seasonal MVP Accuracy: {seasonal_acc*100:.2f}%")
    
    return seasonal_df, seasonal_acc


# --- Tuning RidgeClassifier ---
ridge_params = {
    'alpha': [0.01, 0.1, 1, 10, 50, 100]
}

ridge = RidgeClassifier()
grid_ridge = GridSearchCV(
    ridge,
    ridge_params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)
grid_ridge.fit(X_train_scaled, y_train)

best_ridge = grid_ridge.best_estimator_
print("Best Ridge Params:", grid_ridge.best_params_)

# --- Evaluate tuned Ridge model ---
seasonal_df_ridge, seasonal_acc_ridge = evaluate_ridge_model(
    best_ridge, X_val_scaled, y_val, df_val, "Ridge Regression (Tuned)"
)
```

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
)

# --- Helper Function (Same Format as Ridge) ---
def evaluate_nb_model(model, X_val_scaled, y_val, df_val, model_name):
    """
    Evaluate Naive Bayes model (has predict_proba)
    Output format identical to Ridge Regression evaluation
    """
    y_pred = model.predict(X_val_scaled)
    y_proba = model.predict_proba(X_val_scaled)[:, 1]

    print(f"\nüìä Evaluation Report for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_val, y_pred):.3f} | "
          f"Precision: {precision_score(y_val, y_pred, zero_division=0):.3f} | "
          f"Recall: {recall_score(y_val, y_pred):.3f} | "
          f"F1: {f1_score(y_val, y_pred):.3f} | "
          f"ROC-AUC: {roc_auc_score(y_val, y_proba):.3f}")

    # Attach probabilities for ranking
    df_val = df_val.copy()
    df_val["mvp_prob"] = y_proba
    df_val["pred"] = y_pred

    # --- Pure MVP (Top 5 overall) ---
    top5 = df_val.sort_values("mvp_prob", ascending=False).head(5)
    correct_top5 = top5["is_mvp"].sum()
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top5[["Year", "Player", "mvp_prob", "is_mvp"]])
    print(f"‚úÖ Pure MVP Accuracy (Top-5): {correct_top5}/5 = {correct_top5/5:.2f}")

    # --- Seasonal MVP (per year) ---
    seasonal_preds = []
    for year, group in df_val.groupby("Year"):
        top_player = group.sort_values("mvp_prob", ascending=False).iloc[0]
        seasonal_preds.append({
            "Year": year,
            "Predicted MVP": top_player["Player"],
            "Predicted Prob": top_player["mvp_prob"],
            "Actual MVP?": top_player["is_mvp"]
        })

    seasonal_df = pd.DataFrame(seasonal_preds)
    seasonal_acc = seasonal_df["Actual MVP?"].sum() / len(seasonal_df)

    print("\nüèÄ Seasonal MVP Predictions by Year:")
    print(seasonal_df)
    print(f"\n‚úÖ Seasonal MVP Accuracy: {seasonal_acc*100:.2f}%")

    return seasonal_df, seasonal_acc


# --- Naive Bayes Hyperparameter Tuning ---
nb_params = {'var_smoothing': np.logspace(-12, -3, 10)}

nb = GaussianNB()
grid_nb = GridSearchCV(nb, nb_params, cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)
grid_nb.fit(X_train_scaled, y_train)

best_nb = grid_nb.best_estimator_
print("Best Naive Bayes Params:", grid_nb.best_params_)

# --- Evaluate Tuned Naive Bayes Model ---
seasonal_df_nb, seasonal_acc_nb = evaluate_nb_model(
    best_nb, X_val_scaled, y_val, df_val, "Naive Bayes (Tuned)"
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd

# --- Evaluation identical in style to Ridge ---
def evaluate_rf_model(model, X_val_scaled, y_val, df_val, model_name):
    y_pred  = model.predict(X_val_scaled)
    y_proba = model.predict_proba(X_val_scaled)[:, 1]

    print(f"\nüìä Evaluation Report for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_val,y_pred):.3f} | "
          f"Precision: {precision_score(y_val,y_pred,zero_division=0):.3f} | "
          f"Recall: {recall_score(y_val,y_pred):.3f} | "
          f"F1: {f1_score(y_val,y_pred):.3f} | "
          f"ROC-AUC: {roc_auc_score(y_val,y_proba):.3f}")

    df_val=df_val.copy()
    df_val["mvp_prob"]=y_proba
    df_val["pred"]=y_pred

    # --- Pure MVP (Top 5 overall) ---
    top5=df_val.sort_values("mvp_prob",ascending=False).head(5)
    correct_top5=top5["is_mvp"].sum()
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top5[["Year","Player","mvp_prob","is_mvp"]])
    print(f"‚úÖ Pure MVP Accuracy (Top-5): {correct_top5}/5 = {correct_top5/5:.2f}")

    # --- Seasonal MVP (per year) ---
    preds=[]
    for yr,g in df_val.groupby("Year"):
        tp=g.sort_values("mvp_prob",ascending=False).iloc[0]
        preds.append({"Year":yr,"Predicted MVP":tp["Player"],
                      "Predicted Prob":tp["mvp_prob"],
                      "Actual MVP?":tp["is_mvp"]})
    seasonal_df=pd.DataFrame(preds)
    acc=seasonal_df["Actual MVP?"].sum()/len(seasonal_df)

    print("\nüèÄ Seasonal MVP Predictions by Year:")
    print(seasonal_df)
    print(f"\n‚úÖ Seasonal MVP Accuracy: {acc*100:.2f}%")
    return seasonal_df, acc

# --- Tuning ---
rf_params={
    'n_estimators':[100,300,500],
    'max_depth':[5,10,20,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2']
}
grid_rf=GridSearchCV(RandomForestClassifier(random_state=42,n_jobs=-1),
                     rf_params,cv=3,scoring='roc_auc',n_jobs=-1,verbose=2)
grid_rf.fit(X_train_scaled,y_train)

best_rf=grid_rf.best_estimator_
print("Best Random Forest Params:",grid_rf.best_params_)

# --- Evaluate ---
seasonal_df_rf, seasonal_acc_rf = evaluate_rf_model(
    best_rf, X_val_scaled, y_val, df_val, "Random Forest (Tuned)"
)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def plot_confusion_matrix(model, X_val, y_val, model_name):
    y_pred = model.predict(X_val)
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-MVP', 'MVP'])
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix ‚Äî {model_name} (Baseline)")
    plt.show()

# Example usage (for your baseline models)
plot_confusion_matrix(log_model, X_val_scaled, y_val, "Logistic Regression")
plot_confusion_matrix(ridge_model, X_val_scaled, y_val, "Ridge Regression")
plot_confusion_matrix(nb_model, X_val_scaled, y_val, "Naive Bayes")
plot_confusion_matrix(rf_model, X_val_scaled, y_val, "Random Forest")
```

```python
def plot_confusion_matrix_tuned(model, X_val, y_val, model_name):
    y_pred = model.predict(X_val)
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-MVP', 'MVP'])
    disp.plot(cmap='Greens')
    plt.title(f"Confusion Matrix ‚Äî {model_name} (Tuned)")
    plt.show()

# Example usage (for your tuned models)
plot_confusion_matrix_tuned(best_log, X_val_scaled, y_val, "Logistic Regression (Tuned)")
plot_confusion_matrix_tuned(best_ridge, X_val_scaled, y_val, "Ridge Regression (Tuned)")
plot_confusion_matrix_tuned(best_nb, X_val_scaled, y_val, "Naive Bayes (Tuned)")
plot_confusion_matrix_tuned(best_rf, X_val_scaled, y_val, "Random Forest (Tuned)")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd

def final_evaluation(model, X_val_scaled, y_val, df_val, X_test_scaled, df_test, model_name):
    """
    Evaluate tuned model on validation (2016‚Äì2024)
    and predict MVPs for 2025.
    """
    # --- Validation Performance ---
    y_pred = model.predict(X_val_scaled)
    y_proba = model.predict_proba(X_val_scaled)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(X_val_scaled)
    df_val["mvp_prob"] = y_proba
    df_val["pred"] = y_pred

    acc = accuracy_score(y_val, y_pred)
    prec = precision_score(y_val, y_pred, zero_division=0)
    rec = recall_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)
    roc = roc_auc_score(y_val, y_proba)

    print(f"\nüìä Final Validation Report for {model_name}:")
    print(f"Accuracy: {acc:.3f} | Precision: {prec:.3f} | Recall: {rec:.3f} | F1: {f1:.3f} | ROC-AUC: {roc:.3f}")

    # --- Pure MVP (Top 5 overall) ---
    top5 = df_val.sort_values("mvp_prob", ascending=False).head(5)
    correct_top5 = top5["is_mvp"].sum()
    print("\nüèÜ Top 5 Predicted Pure MVPs (All Years Combined):")
    print(top5[["Year", "Player", "mvp_prob", "is_mvp"]])
    print(f"‚úÖ Pure MVP Accuracy (Top-5): {correct_top5}/5 = {correct_top5/5:.2f}")

    # --- Seasonal MVP Predictions (per year) ---
    results = []
    for year, group in df_val.groupby("Year"):
        top_player = group.sort_values("mvp_prob", ascending=False).iloc[0]
        results.append({
            "Year": year,
            "Predicted MVP": top_player["Player"],
            "Predicted Prob": top_player["mvp_prob"],
            "Actual MVP?": int(top_player["is_mvp"])
        })
    results_df = pd.DataFrame(results)
    print("\nüèÄ Seasonal MVP Predictions by Year:")
    print(results_df)
    print(f"‚úÖ Seasonal MVP Accuracy: {results_df['Actual MVP?'].mean()*100:.2f}%")

    # --- Predict for 2025 ---
    test_proba = model.predict_proba(X_test_scaled)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(X_test_scaled)
    df_test["mvp_prob"] = test_proba
    top5_2025 = df_test.sort_values("mvp_prob", ascending=False).head(5)

    print("\nüîÆ Top 5 Predicted MVPs for 2025 Season:")
    print(top5_2025[["Player", "Team", "Pos", "mvp_prob"]])

    return results_df, top5_2025
```

```python
# Logistic Regression (Tuned)
final_evaluation(best_log, X_val_scaled, y_val, df_val.copy(), X_test_scaled, df_test.copy(), "Logistic Regression (Tuned)")

# Ridge Regression (Tuned)
final_evaluation(best_ridge, X_val_scaled, y_val, df_val.copy(), X_test_scaled, df_test.copy(), "Ridge Regression (Tuned)")

# Naive Bayes (Tuned)
final_evaluation(best_nb, X_val_scaled, y_val, df_val.copy(), X_test_scaled, df_test.copy(), "Naive Bayes (Tuned)")

# Random Forest (Tuned)
final_evaluation(best_rf, X_val_scaled, y_val, df_val.copy(), X_test_scaled, df_test.copy(), "Random Forest (Tuned)")
```

```python
def mvp_2025_predictions(model, X_test_scaled, df_test, model_name):
    """
    Print Seasonal MVP (top 1) and Pure MVP (top 5) for 2025 predictions.
    """
    # Predict probabilities (or decision scores if no predict_proba)
    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_test_scaled)[:, 1]
    else:
        probs = model.decision_function(X_test_scaled)

    df_test = df_test.copy()
    df_test["mvp_prob"] = probs

    # --- Pure MVPs (Top 5 overall for 2025) ---
    top5 = df_test.sort_values("mvp_prob", ascending=False).head(5)
    print(f"\nüîπ {model_name} - Pure MVPs (Top 5 for 2025):")
    print(top5[["Player", "Team", "Pos", "mvp_prob"]])

    # --- Seasonal MVP (Single top player) ---
    top1 = top5.iloc[0]
    print(f"\nüîπ {model_name} - Seasonal MVP for 2025:")
    print(f"{top1['Player']} ({top1['Team']}, {top1['Pos']}) - Predicted Prob: {top1['mvp_prob']:.3f}")


# Example usage for your four models:
mvp_2025_predictions(best_log, X_test_scaled, df_test, "Logistic Regression (Tuned)")
mvp_2025_predictions(best_ridge, X_test_scaled, df_test, "Ridge Regression (Tuned)")
mvp_2025_predictions(best_nb, X_test_scaled, df_test, "Naive Bayes (Tuned)")
mvp_2025_predictions(best_rf, X_test_scaled, df_test, "Random Forest (Tuned)")
```
